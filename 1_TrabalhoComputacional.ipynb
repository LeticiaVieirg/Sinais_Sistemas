a) Determine, de forma analıtica, uma expressão matemática para o sinal x(t) para todo t e plote o sinal x(t).

```python
def x(t):

  y=np.zeros_like(t)
  y[(t>=-1.5) & (t<0)]=2
  y[(t>=0) & (t<=3)]= 2*np.exp(-t[(t>=0) & (t<=3)]/2)

  return y
```

```python
t=np.linspace(-4,9,1000)
xt=x(t)

plt.figure(figsize=(8,3))
plt.plot(t, xt, label='x(t)', linewidth=2)

plt.xlabel('t-->')
plt.ylabel('x(t)')
plt.title('Sinal x(t)')

plt.legend()
plt.grid(True)
plt.show()
```

```python
xti=x(-t)
xtii=x(t+2)
xtiii=x(t-3)
xtiv=x(t*2)
xtv=x(t/3)
```

```python
plt.figure(figsize=(8,3))

plt.subplot(3,2,1)
plt.plot(t, xt, label='x(t)', linewidth=2, color='m')
plt.legend(); plt.grid(True)

plt.subplot(3,2,2)
plt.plot(t, xti, label='x(-t)', linewidth=2, color='m')
plt.legend(); plt.grid(True)

plt.subplot(3,2,3)
plt.plot(t, xtii, label='x(t+2)', linewidth=2, color='m')
plt.legend(); plt.grid(True)

plt.subplot(3,2,4)
plt.plot(t, xtiii, label='x(t-3)', linewidth=2, color='m')
plt.legend(); plt.grid(True)

plt.subplot(3,2,5)
plt.plot(t, xtiv, label='x(2t)', linewidth=2, color='m')
plt.legend(); plt.grid(True)

plt.subplot(3,2,6)
plt.plot(t, xtv, label='x(t/2)', linewidth=2, color='m')
plt.legend(); plt.grid(True)

plt.xlabel('t')
plt.tight_layout()
plt.show()
```
